4.Apply various color manipulation techniques on an image.
(HSV and LAB)


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# Load the image


```python
img = cv2.imread('image.jpg')
```

# Convert to HSV


```python
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

# Convert to LAB


```python
lab_img = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
```

# Convert original image to RGB for display in matplotlib


```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

# For displaying HSV and LAB images properly
# HSV needs scaling since V is in [0,255] but H and S are in different ranges

```python
hsv_img_display = cv2.cvtColor(hsv_img, cv2.COLOR_RGB2HSV)
```

# LAB image may require clipping due to scaling differences


```python
lab_img_display = cv2.cvtColor(lab_img, cv2.COLOR_RGB2LAB)
```

# Display images using matplotlib


```python
plt.figure(figsize=(12, 6))
```

# Show Original Image


```python
plt.subplot(1, 3, 1)
plt.imshow(img_rgb)
plt.title('Original Image')
plt.axis('off')
```

# Show HSV Image (Converted back to RGB for display)


```python
plt.subplot(1, 3, 2)
plt.imshow(hsv_img_display)
plt.title('HSV Image')
plt.axis('off')
```

# Show LAB Image (Converted back to RGB for display)


```python
plt.subplot(1, 3, 3)
plt.imshow(lab_img_display)
plt.title('LAB Image')
plt.axis('off')
```

# Adjust layout


```python
plt.tight_layout()
plt.show()
```

5.Understanding image histograms and applying histogram equalization.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Upload the image file
uploaded = files.upload()

# Load image in grayscale (replace 'image.jpg' with the uploaded filename)
img = cv2.imread(next(iter(uploaded.keys())), 0)  # Load in grayscale

# Calculate histogram
hist = cv2.calcHist([img], [0], None, [256], [0, 256])

# Apply histogram equalization
equalized_img = cv2.equalizeHist(img)

# Display original and equalized images using Matplotlib
plt.figure(figsize=(12, 6))

# Show original image
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Show equalized image
plt.subplot(1, 2, 2)
plt.imshow(equalized_img, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

# Adjust layout
plt.tight_layout()
plt.show()
```

6.Implement various image smoothing and sharpening techniques.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive


# Upload the image file
uploaded = files.upload()

img = cv2.imread(next(iter(uploaded.keys())))


# Gaussian smoothing
smoothed_img = cv2.GaussianBlur(img, (5, 5), 0)

# Sharpening
kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
sharpened_img = cv2.filter2D(img, -1, kernel)


# Display original, smoothed, and sharpened images using Matplotlib
plt.figure(figsize=(15, 5))

# Show original image
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Show smoothed image
plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(smoothed_img, cv2.COLOR_BGR2RGB))
plt.title('Smoothed Image')
plt.axis('off')

# Show sharpened image
plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(sharpened_img, cv2.COLOR_BGR2RGB))
plt.title('Sharpened Image')
plt.axis('off')

# Adjust layout
plt.tight_layout()
plt.show()
```
