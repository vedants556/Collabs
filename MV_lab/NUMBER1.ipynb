```python
# Install OpenCV if not already installed
!pip install opencv-python-headless

import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files



# Load the uploaded image
img = cv2.imread('/content/leh.jpg')

# Check if the image is loaded successfully
if img is None:
    print("Error: Could not load image. Please ensure the file name is correct and the file is uploaded.")
else:
    # Sharpen image using a kernel
    kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
    sharpened_img = cv2.filter2D(img, -1, kernel)

    # Noise removal using Gaussian Blur
    denoised_img = cv2.GaussianBlur(img, (5, 5), 0)

    # Add text watermark
    watermarked_img = img.copy()
    cv2.putText(watermarked_img, 'Watermark', (100, 200), cv2.FONT_HERSHEY_SIMPLEX, 10, (0, 0, 0), 2)

    # Convert BGR to RGB for displaying using matplotlib
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    sharpened_img_rgb = cv2.cvtColor(sharpened_img, cv2.COLOR_BGR2RGB)
    denoised_img_rgb = cv2.cvtColor(denoised_img, cv2.COLOR_BGR2RGB)
    watermarked_img_rgb = cv2.cvtColor(watermarked_img, cv2.COLOR_BGR2RGB)

    # Display images using matplotlib
    plt.figure(figsize=(15, 10))

    # Show Original Image
    plt.subplot(2, 2, 1)
    plt.imshow(img_rgb)
    plt.title('Original Image')
    plt.axis('off')

    # Show Sharpened Image
    plt.subplot(2, 2, 2)
    plt.imshow(sharpened_img_rgb)
    plt.title('Sharpened Image')
    plt.axis('off')

    # Show Denoised Image
    plt.subplot(2, 2, 3)
    plt.imshow(denoised_img_rgb)
    plt.title('Denoised Image')
    plt.axis('off')

   # Show Watermarked Image
    watermarked_img = img.copy()

    # Change text color to white for better visibility
    # The following line was causing the indentation error, it's been fixed to align with the block
    cv2.putText(watermarked_img, 'Watermark', (50, 100), cv2.FONT_HERSHEY_SIMPLEX, 3, (255, 255, 255), 5)
    watermarked_img_rgb = cv2.cvtColor(watermarked_img, cv2.COLOR_BGR2RGB) #converting to rgb for better visibility


    plt.subplot(2, 2, 4)
    plt.imshow(watermarked_img_rgb)
    plt.title('Watermarked Image')
    plt.axis('off')

    plt.show()


```
